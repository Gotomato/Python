```python
X = X.drop("最多風向", 1)
X_test = X_test.drop("最多風向", 1)
X = X.drop(27)
X.head(2)
```

```python
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.utils.validation import check_X_y, check_is_fitted, check_array
import numpy as np

class MyLinearRegression(BaseEstimator, RegressorMixin):
    def __init__(self):
        """
        Initialize a coefficient and an intercept.
        """
        # self.coef_ = -1
        # self.intercept_ = -1
        
    def fit(self, X, y):
        """
        X: data, array-like, shape (n_samples, n_features)
        y: array, shape (n_samples,)
        Estimate a coefficient and an intercept　from data.
        """
        
        """
        self.coef_ = np.linalg.solve(
            np.dot(X.T, X), np.dot(X.T, y)
        )
        """
        X, y = check_X_y(X, y, y_numeric=True)
        
        #切片の計算
        X = np.insert(X, 0, 1, axis=1)
        
        temp = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T),y)
        
        self.coef_ = temp[1:] # 回帰係数
        self.intercept_ = temp[0] # 切片
        return self
    
    def predict(self, X):
        """
        Calc y from X
        """
        check_is_fitted(self, 'coef_') # 学習済みかチェックする(推奨)
        X = check_array(X)
        return np.dot(X,self.coef_) + self.intercept_
```

```python
model = MyLinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X_test)
```
